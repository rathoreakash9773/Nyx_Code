```python
import os
import csv
import requests
from bs4 import BeautifulSoup
import time

# Base directory path where you want to save everything
base_directory = r"/content/Untitled Folder"  # Update this to your desired path

# Create directories if they don't exist
category = 'abstract'
csv_file_name = f'{category}_post_urls.csv'

csv_directory = base_directory
os.makedirs(csv_directory, exist_ok=True)

# CSV file setup
csv_file_name = os.path.join(csv_directory, csv_file_name)

# Create directory for images
image_directory = os.path.join(base_directory, 'images')
os.makedirs(image_directory, exist_ok=True)

# Headers for the request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36',
}

# Function to fetch and process data from the webpage
def fetch_data_from_page(url):
    time.sleep(2)  # Add a delay of 2 seconds between requests
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all posts
        posts = soup.find_all('article', class_='post')

         # Add headings to CSV
        with open(csv_file_name, mode='a', newline='', encoding='utf-8') as file:
            writer = csv.writer(file)
            writer.writerow(['Image URL', 'Title', 'Views'])

        for post in posts:
            # Extract image URL, title, and views
            image_tag = post.find('img', itemprop='image')
            image_url = image_tag['src'] if image_tag else None
            title = image_tag['alt'] if image_tag else None
            views_tag = post.find('div', class_='meta-view')
            views = views_tag.text.strip() if views_tag else None

            if image_url and title and views:
                # Write data to CSV
                with open(csv_file_name, mode='a', newline='', encoding='utf-8') as file:
                    writer = csv.writer(file)
                    writer.writerow([image_url, title, views])

                # Download image
                image_filename = os.path.basename(image_url)
                image_path = os.path.join(image_directory, image_filename)
                if not os.path.exists(image_path):
                    try:
                        image_response = requests.get(image_url, headers=headers)
                        with open(image_path, 'wb') as image_file:
                            image_file.write(image_response.content)
                        print(f"Downloaded image: {image_filename}")
                    except Exception as e:
                        print(f"Failed to download image from URL: {image_url}. Error: {e}")
                else:
                    print(f"Image {image_filename} already exists.")

            else:
                print("Incomplete data for post.")

    else:
        print(f"Failed to fetch page: {url}")

# URL of the webpage to scrape
base_url = "https://negativespace.co/category/abstract/"
num_pages = 1  # Number of pages to scrape

# Iterate through each page
for page_num in range(1, num_pages + 1):
    page_url = base_url if page_num == 1 else f"{base_url}page/{page_num}/"
    print("Constructed URL:", page_url)
    print(f"Scraping page {page_num}...")
    fetch_data_from_page(page_url)

print("Scraping complete.")
```
