```python
import csv
import os
import requests
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver

# Initialize the WebDriver
driver = webdriver.Chrome()
driver.get("https://www.imagesbazaar.com/search/jaipur?sc=1")
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

menu_xpath = "(//div[@class='custom-gallery'])[1]"

i = 1

# Create a CSV file for writing
csv_file_path = "image_data.csv"
with open(csv_file_path, mode='w', newline='') as csv_file:
    fieldnames = ['Image_ID', 'Image_URL']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    
    # Write header row
    writer.writeheader()

    # Create a folder for storing downloaded images
    download_folder = "downloaded_images"
    os.makedirs(download_folder, exist_ok=True)

    while True:
        try:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

            # Wait for the menu to be present
            menu = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, menu_xpath)))

            # Use find_elements instead of find_element to get a list
            main_menus = menu.find_elements(By.XPATH, "(//div[@class='imageItem-Horizontal pie-menu-main-div'])[" + str(i) + "]")

            # Check if the index exists
            if main_menus:
                main_menu = main_menus[0]

                try:
                    # Use a more specific XPath for the image URL
                    url = main_menu.find_element(By.XPATH, ".//img[@class='w-100']")
                    
                    # Get the image URL and extract the Image ID
                    image_url = url.get_attribute("src")
                    image_id = image_url.split("/")[-1].split(".")[0]

                    # Print and write to CSV
                    print(f"{i}: {image_url}, Image ID: {image_id}")
                    writer.writerow({'Image_ID': image_id, 'Image_URL': image_url})

                    # Download and save the image
                    response = requests.get(image_url, stream=True)
                    image_path = os.path.join(download_folder, f"{image_id}.jpg")

                    with open(image_path, 'wb') as image_file:
                        for chunk in response.iter_content(chunk_size=1024):
                            image_file.write(chunk)

                    i += 1
                except Exception as img_exception:
                    print(f"Error getting image URL at index {i}: {str(img_exception)}")
            else:
                # Break the loop if no elements are found
                break

        except Exception as main_exception:
            # Handle specific exceptions if needed
            print(f"rror at index {i}: {str(main_exception)}")
            i += 1

# Close the WebDriver when done
driver.quit()
```
