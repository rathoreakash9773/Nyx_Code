```python

track_urls = df_track_urls['Track URI'].tolist()

track_uris = []
track_names = []
artist_names = []
artist_popularities = []
artist_genres_list = [] 
album_names = []
track_popularities = []
audio_features = []
```

```python
for track_url in track_urls:
    try:
        print(f"Fetching data for track URL: {track_url}")
        track = sp.track(track_url)
        
        track_uri = track["uri"]
        track_uris.append(track_uri)
      
        track_name = track["name"]
        track_names.append(track_name)
       
        artist_uri = track["artists"][0]["uri"]
        artist_info = sp.artist(artist_uri)
       
        artist_name = track["artists"][0]["name"]
        artist_names.append(artist_name)
       
        artist_popularity = artist_info["popularity"]
        artist_popularities.append(artist_popularity)
       
        artist_genres = artist_info["genres"]
        artist_genres_list.append(artist_genres)
        
        album = track["album"]["name"]
        album_names.append(album)
        
        track_popularity = track["popularity"]
        track_popularities.append(track_popularity)
        
        audio_features.append(sp.audio_features(track_uri)[0])
        
    except spotipy.SpotifyException as e:
        print(f"Error fetching data for track URL: {track_url}")
        print(f"Error details: {str(e)}")
```

```python
# Create DataFrame only if audio_features is not empty
if audio_features:
    audio_features_data = pd.DataFrame(audio_features)
else:
    print("No audio features available.")

# Create track_data DataFrame (similar to your existing code)
track_data = pd.DataFrame({
    'Track URI': track_uris,
    'Track Name': track_names,
    'Artist Name': artist_names,
    'Artist Popularity': artist_popularities,
    'Artist Genres': artist_genres_list,
    'Album Name': album_names,
    'Track Popularity': track_popularities
})

# Merge track_data and audio_features_data DataFrames
df_final = pd.merge(track_data, audio_features_data, left_on='Track URI', right_on='uri', how='left')

# Save the final DataFrame to a CSV file
df_final.to_csv("Anwar_merge.csv", index=False)

# Display the final DataFrame
df_final
```
