```python
import pandas as pd

# Read the original CSV file
input_csv = r'C:\Users\Ankit kumar\Desktop\instagram\civitai.csv'
df = pd.read_csv(input_csv)

# Specify the prefix you're looking for
prefix = 'mantine-16xlp3a href'

# Get columns that start with the specified prefix
selected_columns = [col for col in df.columns if col.startswith(prefix)]

# Extract the selected columns and stack them into a single column
stacked_data = df[selected_columns].stack().reset_index(drop=True)

# Create a new DataFrame with the stacked data
result_df = pd.DataFrame({f'{prefix}_data': stacked_data})

# Save the result to a new CSV file
output_csv = 'output_file.csv'
result_df.to_csv(output_csv, index=False)

print(f"Data has been extracted and saved to {output_csv}")
```

```python
import csv
import re
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException, TimeoutException, InvalidArgumentException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Replace 'Your_Web_Driver_Path' with the actual path to your WebDriver executable
driver = webdriver.Chrome()

# Specify the input CSV file path with the URLs
url_csv_file_path = "output_file.csv"

# Specify the output CSV file path for storing extracted data
output_csv_file_path = "extracted_data.csv"

# Open the CSV file containing URLs
with open(url_csv_file_path, 'r') as url_csv_file:
    # Read the URLs from the CSV file
    urls = [row[0] for row in csv.reader(url_csv_file)]

# Iterate through each URL and perform scraping
for url in urls:
    try:
        # Navigate to the webpage
        driver.get(url)

        # Explicitly wait for the title element to be present
        title_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "(//div[@class='mantine-Group-root mantine-1gg8vod'])[6]"))
        )

        # Fetch and print the title
        title = title_element.text
        print(title)

        # Fetch and print the date
        date_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "(//div[@class='mantine-Text-root mantine-1yj1tbi'])[1]"))
        )
        date_tag = date_element.find_element(By.TAG_NAME, "time").get_attribute("datetime")
        print(date_tag)

        # Fetch and print information
        info_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "mantine-9ate7n"))
        )
        info2 = info_element.find_element(By.XPATH, "(//div[@class='mantine-Paper-root mantine-1tnxzwg'])[1]").text

        # Extracting the number of likes and love reactions using regular expressions
        likes_match = re.search(r'ðŸ‘\s*(\d+)', info2)
        love_match = re.search(r'â¤ï¸\s*(\d+)', info2)

        # If matches are found, extract the numbers; otherwise, set to 0
        likes_count = int(likes_match.group(1)) if likes_match else 0
        love_count = int(love_match.group(1)) if love_match else 0

        # Print the counts
        print("Likes:", likes_count)
        print("Love Reactions:", love_count)

        # Store the extracted data in a dictionary
        data = {
            "URL": url,
            "Title": title,
            "Date": date_tag,
            "Likes": likes_count,
            "Love Reactions": love_count
        }

        # Write the data to the output CSV file with proper headers
        with open(output_csv_file_path, 'a', newline='', encoding='utf-8') as output_csv_file:
            # Define the CSV writer
            csv_writer = csv.DictWriter(output_csv_file, fieldnames=data.keys())

            # If it's the first iteration, write headers to the CSV file
            if url == urls[1]:
                csv_writer.writeheader()

            # Write the extracted data to the CSV file
            csv_writer.writerow(data)

    except InvalidArgumentException as e:
        print(f"Invalid URL: {url}. Skipping.")
    except (NoSuchElementException, TimeoutException) as e:
        print(f"Error extracting data from {url}: {e}")
        # Continue to the next URL or handle the error as needed

print(f"Data has been successfully written to {output_csv_file_path}")

# Close the WebDriver
driver.quit()
```

```python
import csv
from bs4 import BeautifulSoup
import requests

# Function to extract image URL from the webpage
def extract_image_url(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    image_url = soup.select_one('.mantine-it6rft')['src']
    return image_url

# Assuming 'urls.csv' is your CSV file with a column named "ImageURL"
csv_input_filename = 'output_file.csv'
csv_output_filename = 'image_urls.csv'

with open(csv_input_filename, 'r') as csv_input_file, open(csv_output_filename, 'w', newline='') as csv_output_file:
    csv_reader = csv.DictReader(csv_input_file)
    fieldnames = ['mantine-16xlp3a href_data']

    # Create CSV writer
    csv_writer = csv.DictWriter(csv_output_file, fieldnames=fieldnames)
    csv_writer.writeheader()

    for row in csv_reader:
        url = row['mantine-16xlp3a href_data']

        # Extract image URL using BeautifulSoup
        try:
            image_url = extract_image_url(url)

            # Write the image URL to the output CSV file
            csv_writer.writerow({'mantine-16xlp3a href_data': image_url})
            print(f"Image URL extracted successfully from {url}")
        except Exception as e:
            print(f"Error extracting image URL from {url}: {e}")

print(f"Image URLs extracted and saved in {csv_output_filename}")
```

```python
import os
import csv
import requests
from urllib.parse import urlparse
from concurrent.futures import ThreadPoolExecutor

def download_image(url, destination_folder):
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # Raise an HTTPError for bad responses
    except requests.exceptions.RequestException as e:
        print(f"Failed to download {url}. Error: {e}")
        return

    # Extract the file name from the URL
    file_name = urlparse(url).path.split("/")[-1]
    file_path = os.path.join(destination_folder, file_name)

    # Save the image to the destination folder
    try:
        with open(file_path, 'wb') as file:
            file.write(response.content)
        print(f"Downloaded: {file_path}")
    except IOError as e:
        print(f"Error processing {url}. Error: {e}")

def download_images_from_csv(csv_file, destination_folder):
    with open(csv_file, 'r') as file:
        reader = csv.reader(file)

        # Create the destination folder if it doesn't exist
        os.makedirs(destination_folder, exist_ok=True)

        # Use ThreadPoolExecutor for concurrent downloading
        with ThreadPoolExecutor() as executor:
            for row in reader:
                if len(row) > 0:
                    image_url = row[0]  # Assuming the image URL is in the first column
                    executor.submit(download_image, image_url, destination_folder)
                else:
                    print("Skipping row - empty:", row)

if __name__ == "__main__":
    csv_file_path = r"image_urls.csv"  # Replace with the path to your CSV file
    output_folder = r"civit"  # Replace with your desired location

    download_images_from_csv(csv_file_path, output_folder)
```
