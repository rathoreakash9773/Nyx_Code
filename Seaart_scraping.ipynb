```python
import os
import requests
from bs4 import BeautifulSoup

# Set the output directory
output_dir = r'D:\Python\NYX\Insta_folders\Output'
os.makedirs(output_dir, exist_ok=True)

url = 'https://www.seaart.ai/tagInfo/photography'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

image_links = []
for img in soup.find_all('img'):
    src = img.get('src')
    if src.startswith('http'):
        image_links.append(src)

for i, link in enumerate(image_links, 1):
    print(f'Downloading image {i}: {link}')
    response = requests.get(link)
    file_name = os.path.join(output_dir, f'image_{i}.jpg')
    with open(file_name, 'wb') as file:
        file.write(response.content)

print('Image download completed.')
```

```python
import os
import requests
from bs4 import BeautifulSoup
import csv

# Set the output directory
output_dir = r'D:\Python\NYX\Insta_folders\Output'
os.makedirs(output_dir, exist_ok=True)

url = 'https://www.seaart.ai/tagInfo/realisticall'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

image_links = []
for img in soup.find_all('img'):
    src = img.get('src')
    if src.startswith('http'):
        image_links.append(src)

# Create a CSV file for storing information
csv_file_path = os.path.join(output_dir, 'image_info.csv')
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['ImageNumber', 'ImageURL', 'FileName']
    csv_writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    # Write header to CSV file
    csv_writer.writeheader()

    # Download images and write information to CSV
    for i, link in enumerate(image_links, 1):
        print(f'Downloading image {i}: {link}')
        response = requests.get(link)
        file_name = os.path.join(output_dir, f'image_{i}.jpg')

        # Write information to CSV
        csv_writer.writerow({'ImageNumber': i, 'ImageURL': link, 'FileName': file_name})

        # Save image file
        with open(file_name, 'wb') as file:
            file.write(response.content)

print('Image download and CSV creation completed.')
```
