# Civit

```python
import requests
import urllib.parse
import json

# Define the base URL for the API endpoint
url = 'https://civitai.com/api/trpc/image.getInfinite'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36',
    'Referer':'https://civitai.com/images'
}
# Initialize the payload with the cursor set to None
payload = {
    "json": {
        "period": "Week",
        "sort": "Most Reactions",
        "types": ["image"],
        "cursor": None  # Initial cursor value
    },
    "meta": {
        "values": {
            "cursor": ["undefined"]
        }
    }
}

page = 0
# Loop until no more data is returned
while True:
    # URL encode the payload
    encoded_payload = urllib.parse.quote(json.dumps(payload))
    full_url = f"{url}?input={encoded_payload}"

    # Make the request
    response = requests.get(full_url, headers=headers)

    if response.status_code == 200:
        page += 1
        data = response.json()
        items = data['result']['data']['json']['items']
        print('page:',page, len(items))  # Placeholder for actual data processing

        # Determine the next cursor from the response
        next_cursor = data.get('result').get('data').get('json').get('nextCursor')
        if page % 200 == 0:
            print('page:',page,'cursor value:',next_cursor)

        # if next_cursor == 'desired_cursor_for_custom_lastpage' -> break
        if not next_cursor:
            print("Reached the last page.")
            break

        # Update the cursor in the payload for the next iteration
        payload['json']['cursor'] = next_cursor
    else:
        print("Failed to fetch data:", response.status_code)
        break
```
